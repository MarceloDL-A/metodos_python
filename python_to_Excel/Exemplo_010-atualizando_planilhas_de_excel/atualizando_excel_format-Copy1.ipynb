#### Carregando uma Planilha Existente

```python
from openpyxl import load_workbook
import pandas as pd
caminho = 'mascara_excel_v002.xlsx'
arquivo_excel = load_workbook(caminho)
```

#### Composição de um dataframe com os dados do excel

#### Atualização das planilhas excel com os arquivos do dataframe

##### Descrição das sheets

```python
for sheet in arquivo_excel:
    print("\n")
    print(sheet)
    for row in sheet.rows:
        print("\n")
        for cell in row:
            try:
                print(cell, cell.coordinate, cell.col_idx)
            except:
                pass
```
