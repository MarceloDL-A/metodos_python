Um jogador paga 5 reais para participar de um jogo de dados, disputando com o oponente quem tem o ponto maior. O jogador e o oponente lançam cada um o seu dado e a seguinte regra de premiação é estabelecida:<br>
* Se o ponto do jogador é maior, ele ganha 2 vezes a diferença entre o seu ponto e o obtido pelo oponente.<br>
* Se o ponto do jogador é menor ou igual ao do oponente, ele não ganha nada. <br>
O que você acha desse jogo? Vale a pena para o jogador? 

Vamos simular 100 jogos e calcular a probabilidade de lucro, que é ganhar mais 6 reais ou mais no jogo.

```python
from random import seed
from random import randint
# semente do gerador de números aleatórios
seed(1)

aposta = 5 # Em cada rodada deve ser apostado 5 reais
Ng = 100 # numero de jogos
Ns = 10
p = 0
for n in range(1,Ng):
    G = 0
    j = randint(1, 6) # resultado do dado do jogador
    b = randint(1, 6) # resultado do dado do oponente
    if(j > b): # se o valor do jogador é maior
        G = 2*(j-b) # ganho em uma única jogada
    if(G > aposta): # se houver lucro
        p = p + 1
p = p/Ns # probabilidade de ter lucro
print('Aposta:', aposta, 'reais.')
print("Chance de vitória:", p)
print("Valor teórico:", 1/6)
```

É possível ainda verificarmos como a chance de lucro muda com o valor da aposta. 

```python
from matplotlib import pyplot as plt
import numpy as np
seed(100)

plt.figure(figsize=(10,6))
vp = [] # guarda o valor apostado
vaposta = [] # guarda a probabilidade de lucro
for aposta in np.arange(0,11,2):
    Ng = 100 # numero de jogos
    p = 0
    for n in range(1,Ng):
        G = 0 # ganho
        j = randint(1, 6) # valor do jogados
        b = randint(1, 6) # valor do oponente
        if(j > b): # se o valor do jogador é maior
            G = 2*(j-b)
        if(G > aposta): # se houver lucro
            p = p + 1
    p = p/Ng # probabilidade de ter lucro
    vp.append(p)
    vaposta.append(aposta)
plt.plot(vaposta, vp, 'ro-', label = 'Simulacao')
plt.xlabel('Valor da aposta', fontsize = 15)
plt.ylabel('Probabilidade de lucro',fontsize = 15)

# valor teorico
vpt = [15/36,10/26,6/36,3/36,1/36,0] #prob. teorica de lucro
vapostat=[0,2,4,6,8,10] # aposta
plt.plot(vapostat, vpt, 'bo-',label='Valor teórico')
plt.legend(fontsize = 15)
plt.savefig('Figure.pdf')
plt.show(True)
```
